## Fourier Analysis

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Creating a time-domain signal

```python
fs = 1500 # sampling frequency
T = 3.0 # total duration
'''
t creates evenly spaced intervals from 0 to T with total intervals being frequency x T
'''
t = np.linspace(0, T, int(fs*T), endpoint = False) # Time vector

'''
signal definition - here we a forming a signal that oscillates at a frequency f = signal = sin(2Ï€ft),
this is one of the mathematical representation of signals
'''
signal = np.sin(2 * np.pi * 10 * t)  # A 10 Hz sine wave

'''
Here we are plotting the signal 
'''

def plotSignal(vector, signal, freq):
    plt.plot(t, signal)
    plt.xlabel("Time (s)")
    plt.ylabel("Amplitude")
    plt.title(f"{freq} Hz Sine Wave")
    plt.show()

plotSignal(t, signal, "10")

# Composite signal: sum of two sine waves (10 Hz and 50 Hz)
freq1 = 10  # Frequency of first sine wave (Hz)
freq2 = 50  # Frequency of second sine wave (Hz)

'''
Here we are forming a complex signal made of two signals, by now you understand the np.sin(2 * np.pi * freq * t)
The extra parameter we have added is the 2 * and the 1 *, which represents amplitudes, the first freq is
operating in an amplitude of 2, and the second wave is operating in an amplitude of 1.
'''
signal = 2 * np.sin(2 * np.pi * freq1 * t) + 1 * np.sin(2 * np.pi * freq2 * t)
plotSignal(t, signal, "composite")
```


In time domain, a signal represents how a quantity (such as voltage, sound) varies over time. Let us take for example ECG (Electrocardiogram signal), you see the waves you see in movies before someone dies, the amplitude represents the strength of the electrical signal or mechanical activity associated with each heartbeat. That is why when a person dies it flattens out. It is a time domain signal and at that particular time, when the amplitude zeros out, the heart has stopped working.

The __ECG__ is actually a  perfect example of a composite signal. The amplitude represents the voltage generated by electrical impulses as the heart relaxes and contracts. Each heartbeat produces a characteristic waveform due to this electrical activity in the heart which can be broken down into:

1. __P wave__ - atrial depolarization, the electrical impulse triggering the atria to contract
2. __QRS complex__ - ventricular depolarization - the electrical impulse triggering the ventricles to contract
3. __T wave__ - ventricular repolarization - the recovery phase of the ventricles.


The larger the amplitude the stronger the electrical activity in the heart
