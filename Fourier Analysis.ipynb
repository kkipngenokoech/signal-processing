# SIGNAL PROCESSING

## Fourier Analysis

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Creating a time-domain signal

```python
fs = 500 # sampling frequency
T = 1.0 # total duration
t = np.linspace(0, T, int(fs*T), endpoint = False) # Time vector

signal = np.sin(2 * np.pi * 10 * t)  # A 10 Hz sine wave

def plotSignal(vector, signal, freq):
    plt.plot(t, signal)
    plt.xlabel("Time (s)")
    plt.ylabel("Amplitude")
    plt.title(f"{freq} Hz Sine Wave")
    plt.show()

plotSignal(t, signal, "10")

# Composite signal: sum of two sine waves (10 Hz and 50 Hz)
freq1 = 10  # Frequency of first sine wave (Hz)
freq2 = 50  # Frequency of second sine wave (Hz)
signal = 2 * np.sin(2 * np.pi * freq1 * t) + 1 * np.sin(2 * np.pi * freq2 * t)
plotSignal(t, signal, "composite")
```
